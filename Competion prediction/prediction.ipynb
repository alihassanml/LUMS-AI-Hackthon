```python
import os
import pandas as pd
import numpy as np
from tensorflow.keras.models import load_model

DATA_DIR = 'Final Dataset'  
NUM_FRAMES = 100
MODEL_PATH = 'model_1.h5' 

model = load_model(MODEL_PATH)

results = []

def load_and_process_data(file_path):
    data = pd.read_csv(file_path)
    if len(data) == NUM_FRAMES:  # Check if the data has the required number of frames
        return data[['x', 'y']].values  # Extract 'x' and 'y' columns as a NumPy array
    return None

# Sort files in sequence based on their name
file_names = sorted(os.listdir(DATA_DIR), key=lambda x: int(x.split('.')[0]))  # Sorting by the numeric part of the file name

for file_name in tqdm(file_names):
    file_path = os.path.join(DATA_DIR, file_name)
    if file_path.endswith('.csv'):  # Process only CSV files
        input_data = load_and_process_data(file_path)
        if input_data is not None:
            # Reshape input data to match model input
            input_data = np.expand_dims(input_data, axis=0)  # Add batch dimension
            predictions = model.predict(input_data)  # Get predictions
            predicted_class = np.argmax(predictions, axis=1)  # Get class index
            one_hot_encoded = np.eye(predictions.shape[1])[predicted_class]  # One-hot encode
            
            # Convert one-hot encoded vector to integer values
            one_hot_encoded_int = one_hot_encoded.astype(int)  # Convert to integer
            
            # Append results as a dictionary with integer labels
            results.append({
                'file_name': int(file_name.split('.')[0]),
                'labels': one_hot_encoded_int.flatten().tolist()  # Flatten and ensure integer values
            })

# Create DataFrame from results
results_df = pd.DataFrame(results)

# Save results to a CSV file
results_df.to_csv('result_1.csv', index=False)

print("Results saved to result3.csv")
```
